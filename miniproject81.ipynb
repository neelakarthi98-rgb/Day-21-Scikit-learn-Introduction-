## Basic Perceptron Classifier

//**Objective: Use sklearn.linear_model.Perceptron to classify a simple binary dataset.**

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

//Generate Data:

```python
X,y=make_classification(n_samples=200,n_features=2,n_informative=2,n_redundant=0,n_clusters_per_class=1,random_state=42)
```

//Split data:

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
model=Perceptron(max_iter=1000,eta0=1.0,random_state=42)
model.fit(X_train,y_train)
```

//make predictions//

```python
y_pred=model.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print("Model Accuracy:",accuracy)
```

//visualization//

```python
plt.figure(figsize=(8,6))
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, cmap='coolwarm', edgecolor='k', s=60)
plt.title("Perceptron Classification (Predicted Labels)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
